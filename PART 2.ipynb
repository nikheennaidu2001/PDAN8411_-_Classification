### KNN Model

#### Training and Testing with KNN

We split the dataset into two pieces, one for training and the other for testing, using the X and Y variables we just declared.

```python
# FIRST WE IMPORT THE train_test_split LIBRARY
from sklearn.model_selection import train_test_split # THIS IMPORT IS USED IN SPLITING THE DATA FOR TESTING AND TRAINING

# THIS SPLITS THE DATA, test_size=0.3 OR 30%, AND THE TRAIN SIZE IS EQUAL TO THE REMAINING 70%
# WHY random_state=42, IS TO ENSURE THAT THE OUTPUT THAT IS RUN EACH TIME EQUALS TO THE FIRST RUN OUTPUT WHEN SPLITTING
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

Next we scale our data. The reason why we scale data into new values because it makes it easier to compare.

```python
# FIRST WE IMPORT THE StandardScaler LIBRARY
from sklearn.preprocessing import StandardScaler # THIS IMPORT IS USED FOR SCALING DATA

# DECLARE THE scaler OBJECT AND VARIABLES FOR SCALING THE X_train & X_test
scaler = StandardScaler()
scaled_X_train = scaler.fit_transform(X_train) # VARIABLE FOR X_train
scaled_X_test = scaler.transform(X_test) # VARIABLE FOR X_test
```

Than we bring in the KNN model. We use KNN to predict the target variable using one or multiple independent variables.

```python
# IMPORT THE KNeighborsClassifier LIBRARY
from sklearn.neighbors import KNeighborsClassifier # THIS IMPORT IS USED FOR THE KNN MODEL

# DECLARE A VARIABLE FOR THE KNN OBJECT
knn_model = KNeighborsClassifier(n_neighbors=3) # K=3 BECAUSE THAT IS WHERE THE MODEL IS MOST OPTIMISED
knn_model.fit(scaled_X_train, Y_train) # FIT THE KNN MODEL WITH THE scaled_X_train & Y_train
```

#### Predictions with KNN

We make predictions based on our data

```python
# WE FIRST DECLARE A VARIABLE FOR PREDICTIONS USING THE SCALED X_test
Y_pred = knn_model.predict(scaled_X_test)
```

Test these Predictions

```python
# FIRST WE IMPORT THE confusion_matrix & classification_report LIBRARIES
from sklearn.metrics import confusion_matrix, classification_report

# DECLARING THE confusion_matrix OBJECT USING Y_test & Y_pred
confusion_matrix(Y_test, Y_pred)
```

We use the len() function to see the full length of Y_test

```python
# len() RETURNS THE LENGTH OF AN OBJECT
len(Y_test)
```

#### Checking the Accuracy of the Model with KNN

The classification_report() is used to measure the quality of predictions from a classification algorithm.

```python
# PRINTS THE ACCURACY OF THE KNN MODEL USING Y_test & Y_pred
print(classification_report(Y_test, Y_pred))
```

The value_counts() method returns a count of unique values in a specific column or dataset.

#### Accuracy Score and Error Rate with KNN

We use the accuracy score by dividing the number of correct predictions by the total prediction number

```python
# WE IMPORT THE accuracy_score LIBRARY
from sklearn.metrics import accuracy_score # USED FOR accuracy_score

# WE DECLARE A VARIABLE FOR ERROR RATES
test_error_rates = []

for i in range(1,15): # RANGE FOR N IS BETWEEN 1 TO 15
    knn_model = KNeighborsClassifier(n_neighbors=i)
    knn_model.fit(scaled_X_train, Y_train)

    Y_pred_test = knn_model.predict(scaled_X_test) 
    test_error = 1 - accuracy_score(Y_test, Y_pred_test)
    test_error_rates.append(test_error)
```

The test_error_rates method is used to prediction error of a model made with respect to the true model.

```python
# DISPLAY THE test_error_rates VALUES
test_error_rates
```

Than finally we visualise the test error rates on a graph

```python
# DISPLAY THE RANGE OF 1 TO 15
plt.plot(range(1,15), test_error_rates)
# Y LABEL
plt.ylabel('Error Rate')
# X LABEL
plt.xlabel('K neighbors')
# PRINTS GRAPH
plt.show()
```

#### GridSearch Cross Validation with KNN

We use GridSearch cross validation to select the best of the machine learning model, parameterized by a grid of hyperparameters.

```python
# WE FIRST IMPORT THE GridSearchCV LIBRARY
from sklearn.model_selection import GridSearchCV 

# SET UP K-NEIGHBOURS AS A LIST
k_values = list(range(1,5)) # RANGE IS 5, AS OF MOMENT THE MOST OPTIMISED K=NEIGHBOUR IS 5
# PRINT THEM OUT
k_values
```

Pass these as parameters  

```python
# WE USE THE VARAIBLE 'k_values' DECLARED ABOVE INTO THE PARAMETERS
param_grid = {'knn__n_neighbors': k_values}
```

Before we setting up our classifier, we first must setup the operations of the pipeline

```python
# WE FIRST IMPORT THE Pipeline LIBRARY
from sklearn.pipeline import Pipeline

# SETUP THE 'operations' FOR THE PIPELINE
operations = [('scaler', scaler),('knn',knn_model)]

# PASS THE 'operations' TO THE PIPELINE
pipe = Pipeline(operations)
```

Setting up our classifier 

```python
# PASSING IN THE PIPELINE (WITH OPERATIONS), PARAMETERS, A CLASSIFIED VALUE, & CRITERIA FOR SCORING
full_cv_classifier = GridSearchCV(pipe, param_grid, cv = 5, scoring='accuracy')
```

Then we train the cross validation model

```python
# USING THE X_train & Y_train
full_cv_classifier.fit(X_train, Y_train)
```

Get the parameters of the best estimate of classifier

```python
# GET THE PARAMETERS FOR BEST K-NEIGHBOURS
full_cv_classifier.best_estimator_.get_params()
```

Testing

```python
# WE USE THE X_test TO COMPARE DATA
full_pred = full_cv_classifier.predict(X_test)
```

Than we compare the predicted values to actual values then determine accuracy

```python
# FROM THIS WE LOOK AT THE 'precision' OF EACH WITH AN ACCURACY OF 0.95%
print(classification_report(Y_test, full_pred))
```

Mean square errors of the predictions

Than we get the mean square error
